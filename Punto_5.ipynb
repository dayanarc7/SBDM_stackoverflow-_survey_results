## NUEVOS DATOS PARA EL 2018 DE LOS LENGUAJES DE PROGRAMACIÓN

```python
#Cargue de librerias

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams

from sklearn.neural_network import MLPRegressor
```

```python
#Cargue de archivo como un dataframe

ruta_origen = 'C:/Users/Marvin/Desktop/Bases de datos masivas/Trabajo Final/dw_stackoverflow_.csv'
df=pd.read_csv(ruta_origen,delimiter=';')

df.replace("null", 0)
nombre_columnas = df.columns.tolist()
nombre_columnas

#Crear Dataframe con lenguajes de programacion
df1 = df[nombre_columnas[0:14]]

#Reemplazar los id del tiempo por los años
df1['fk_tiempo'] = df1['fk_tiempo'].map({1: '2013',2: '2016',3: '2014',4: '2011',5: '2012',6: '2015'})
```

## Proceso para proyectar para el año 2017, los campos asociados a los lenguajes de programación y Ocupación

```python
#Total de encuestas
df1_tot = df1.groupby(['fk_tiempo'])[nombre_columnas[0:13]]
df1_tot = df1_tot.agg(np.size)
df1_tot
```

```python
#Total de encuestas diligenciadas
nombre_columnas = df1.columns.tolist()
df1_filtro = df1.groupby(['fk_tiempo'])[nombre_columnas[0:13]].sum()
df1_filtro
```

```python
#Porcentaje del lenguaje seleccionado respecto al total de encuestas
df1_porc= round(df1_filtro*100/df1_tot,0)
df1_porc
```

```python
#Grafico - Comportamiento desde 2011 a 2016
fig=df1_porc.plot()
rcParams['figure.figsize'] =15, 5
plt.title('Porcion de lenguaje de programación por año')
plt.xlabel('Años')
plt.ylabel('Porcentaje')
```

```python
# Porcentaje a aplicar en 2017 mediante Promedio movil con m= 3
y = round (df1_porc[4:8].mean(),0)
y
```

```python
# Crear nuevos datos para 2017
y = round (df1_porc[4:8].mean(),0)
cantidad = 50000
columnas=  len(y)

p1=int(y[0:columnas][0])/100
p2=int(y[0:columnas][1])/100
p3=int(y[0:columnas][2])/100
p4=int(y[0:columnas][3])/100
p5=int(y[0:columnas][4])/100
p6=int(y[0:columnas][5])/100
p7=int(y[0:columnas][6])/100
p8=int(y[0:columnas][7])/100
p9=int(y[0:columnas][8])/100
p10=int(y[0:columnas][9])/100
p11=int(y[0:columnas][10])/100
p12=int(y[0:columnas][11])/100
p13=int(y[0:columnas][12])/100

year =  ["2017"] * cantidad
j = [1] * int(p1 * cantidad) + [None] * (cantidad-int(p1 * cantidad))
js = [1] * int(p2 * cantidad) + [None] * (cantidad-int(p2 * cantidad))
p = [1] * int(p3 * cantidad) + [None] * (cantidad-int(p3 * cantidad))
y = [1] * int(p4 * cantidad) + [None] * (cantidad-int(p4 * cantidad))
r = [1] * int(p5 * cantidad) + [None] * (cantidad-int(p5 * cantidad))
q = [1] * int(p6 * cantidad) + [None] * (cantidad-int(p6 * cantidad))
c = [1] * int(p7 * cantidad) + [None] * (cantidad-int(p7 * cantidad))
c2 = [1] * int(p8 * cantidad) + [None] * (cantidad-int(p8 * cantidad))
d = [1] * int(p9 * cantidad) + [None] * (cantidad-int(p9 * cantidad))
f = [1] * int(p10 * cantidad) + [None] * (cantidad-int(p10 * cantidad))
t = [1] * int(p11 * cantidad) + [None] * (cantidad-int(p11 * cantidad))
e = [1] * int(p12 * cantidad) + [None] * (cantidad-int(p12 * cantidad))
w = [1] * int(p13 * cantidad) + [None] * (cantidad-int(p13 * cantidad))


#Crear el data frame
df1_nuevo = pd.DataFrame({"lenguaje_java": j, "lenguaje_javascript": js, "lenguaje_php": p,"lenguaje_python": y,"lenguaje_ruby": r,"lenguaje_sql": q, "lenguaje_c#": c, "lenguaje_c++": c2, "ocupacion_Desktop": d,"ocupacion_Staff": f,"ocupacion_Database": t,"ocupacion_Embedded": e,"ocupacion_Web": w,"fk_tiempo": year})

df1_nuevo = df1_nuevo.fillna(value=0)

#Ver una muestra de la nueva tabla
df1_nuevo.head(10)
```

```python
# Ensamble de nuevos datos a la bodega
df1_union = df1.append(df1_nuevo)
df1_union.head(10)
```

```python
#Total de encuestas
dfu1_tot = df1_union.groupby(['fk_tiempo'])[nombre_columnas[0:13]]
dfu1_tot = dfu1_tot.agg(np.size)

#dfu1_tot
nombre_columnas = df1_union.columns.tolist()
dfu1_filtro = df1_union.groupby(['fk_tiempo'])[nombre_columnas[0:13]].sum()
dfu1_filtro
dfu1_porc= round(dfu1_filtro*100/dfu1_tot,0)
dfu1_porc

```

## Proceso para proyectar para el año 2017, los campos asociados a la ubicación, rango de ingresos y Experiencia.

```python
#Para la Experiencia


df2 = df[['fk_tiempo','fk_experiencia']]

df2['fk_tiempo'] = df2['fk_tiempo'].map({1: '2013',2: '2016',3: '2014',4: '2011',5: '2012',6: '2015'})

df2['fk_experiencia'] = df2['fk_experiencia'].map({1: '6-10',2: '<2',3: '11',4: '2-5'})

#Tabla con comportamiento de porcentaje relativo

df3=pd.crosstab(index=df2['fk_tiempo'],
            columns=df2['fk_experiencia'], margins=False).apply(lambda r: r/r.sum() *100,
                                axis=1)
df3
```

```python
# Grafico para observar proporcion en cada año por rango de años de experiencia

plot = pd.crosstab(index=df2['fk_tiempo'],
            columns=df2['fk_experiencia']).apply(lambda r: r/r.sum() *100,
                                              axis=1).plot(kind='bar', stacked=True)
```
